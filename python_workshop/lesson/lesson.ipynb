```python
## Analyzing Data from Multiple Files

import glob
import numpy as np

# Use glob to find all inflammation data files
file_pattern = '../data/inflammation-*.csv'  # Adjust the path to match your folder structure
all_files = glob.glob(file_pattern)

# Initialize an empty list to store data arrays
data_arrays = []

# Loop through all matching files and load them into NumPy arrays
for file in all_files:
    print(file)
    data = np.loadtxt(file, delimiter=',')
    data_arrays.append(data)

## Selecting the first four datasets for analysis and visualization
sample=data_arrays


for datasets in sample:

    ## Mean/max/min inflammation per day per file
    mean_perday=np.mean(datasets, axis=0)
    max_perday=np.amax(datasets, axis=0)
    min_perday=np.amin(datasets, axis=0)

    ## Creating the figure of three subplots

    fig=plt.figure(figsize=(15.0, 3.0))

    ## Plotting in subplots
    axes1=fig.add_subplot(1, 3, 1)
    axes2=fig.add_subplot(1, 3, 2)
    axes3=fig.add_subplot(1, 3, 3)

    axes1.plot(mean_perday)
    axes2.plot(max_perday)
    axes3.plot(min_perday)

    ##### add this part later to make the plots more readable####
    ## Naming axis that are different for each plot
    axes1.set_ylabel('mean')
    axes2.set_ylabel('max')
    axes3.set_ylabel('min')

    ## Naming axis that are the same for each plot
    for ax in [axes1,axes2,axes3]:
        ax.set_xlabel('day')


#### Do it after inspecting the first plots######
## Close inspection at the third dataset that has zero for each day for all patients

third_dataset=data_arrays[2]

fig=plt.figure(figsize=(15.0, 3.0))
plt.imshow(third_dataset)
plt.colorbar()
plt.title('Inflammation dataset 3')
plt.xlabel('day')
plt.ylabel('patient')

#interpretation: 
# - there are some zero values across all days for every patient so that might be an error in the data collection process
# - the last patient seems not having any inflammation at all, so maybe it is not sick ...(patient to follow up with)
```

```python
# Situation : what if we want to convert our data to a different scale? like from farhenheit to celsius

## For one data value (first):

fahrenheit_value=99
celsius_value=(fahrenheit_value-32)*(5/9)

print(celsius_value)


## for a list of data values (second):

fahrenheit_values=[99, 100, 101]
celsius_values=[]
for fahrenheit_value in fahrenheit_values:
    celsius_value=(fahrenheit_value-32)*(5/9)
    celsius_values.append(celsius_value)

print(celsius_values)
```

```python
## But, if we would like to:
# - reuse the code without copying it again 
# - reduce redundancy 
# - make the code more readable
#- avoid state dependencies (i.e. the code depends on the variable values elsewehere in the code)


## We can define a function that takes the fahrenheit value and returns the celsius value (third)

## Explain the function definition (def + function name + arguments(parameter names) + return statement+return value)

def fahrenheit_to_celsius(fahrenheit_value):
    celsius_value=(fahrenheit_value-32)*(5/9)
    return celsius_value

## Explain the function call (function name + arguments) (fourth)

fahrenheit_to_celsius(99)

### Calling the function for other values (fifth)

print("freezing point of water:", fahrenheit_to_celsius(32),"C")
print("boiling point of water:", fahrenheit_to_celsius(212),"C")
```

## Composing functions 

```python
## We can compose functions

## Example write a function that takes farenheit and converted to kelvin by two functions, one to convert to celsius and the other to convert to kelvin

def celsius_to_kelvin(celsius_value):
    kelvin_value=celsius_value+273.15
    return kelvin_value

def fahrenheit_to_kelvin(fahrenheit_value):
    celsius_value=fahrenheit_to_celsius(fahrenheit_value)
    kelvin_value=celsius_to_kelvin(celsius_value)
    return kelvin_value

print("freezing point of water in Kelvin:", fahrenheit_to_kelvin(32),"K")
print("boiling point of water in Kelvin:", fahrenheit_to_kelvin(212),"K")
```


######### Go to slides to summarize defensive programming ##############

Broadly speaking, assertions fall into three categories:

A precondition is something that must be true at the start of a function in order for it to work correctly.

A postcondition is something that the function guarantees is true when it finishes.

An invariant is something that is always true at a particular point inside a piece of code.

## OPTIONAL!!! Debugging (for presentation)

### Test with simplified data. 
Before doing statistics on a real data set, we should try calculating statistics for a single record, for two identical records, for two records whose values are one step apart, or for some other case where we can calculate the right answer by hand.

### Test a simplified case. 
If our program is supposed to simulate magnetic eddies in rapidly-rotating blobs of supercooled helium, our first test should be a blob of helium that isn’t rotating, and isn’t being subjected to any external electromagnetic fields. Similarly, if we’re looking at the effects of climate change on speciation, our first test should hold temperature, precipitation, and other factors constant.

### Compare to an oracle. 
A test oracle is something whose results are trusted, such as experimental data, an older program, or a human expert. We use test oracles to determine if our new program produces the correct results. If we have a test oracle, we should store its output for particular cases so that we can compare it with our new results as often as we like without re-running that program.

### Check conservation laws.
Mass, energy, and other quantities are conserved in physical systems, so they should be in programs as well. Similarly, if we are analyzing patient data, the number of records should either stay the same or decrease as we move from one analysis to the next (since we might throw away outliers or records with missing values). If “new” patients start appearing out of nowhere as we move through our pipeline, it’s probably a sign that something is wrong.

### Visualize. 
Data analysts frequently use simple visualizations to check both the science they’re doing and the correctness of their code (just as we did in the opening lesson of this tutorial). This should only be used for debugging as a last resort, though, since it’s very hard to compare two visualizations automatically.


## OPTIONAL!!!! Strategy debugging

### Change one thing at a time

Replacing random chunks of code is unlikely to do much good. (After all, if you got it wrong the first time, you’ll probably get it wrong the second and third as well.) Good programmers therefore _change one thing at a time, for a reason_.

- Gather more information: "Is the bug still there?"
- Test a fix: "Can we make the bug go away?"

### Keep track of what you've done

Good scientists keep track of what they’ve done so that they can reproduce their work. People are more likely to listen to us when we can explain clearly what we did, and we’re better able to give them the information they need.

Version control is often used to reset software to a known state during debugging, and to explore recent changes to code that might be responsible for bugs. More on this during the lesson on version control!
